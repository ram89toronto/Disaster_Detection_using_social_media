```python
import pandas as pd
import snscrape.modules.twitter as sntwitter
import itertools
```

### Fire:

```python
df = pd.DataFrame(itertools.islice(sntwitter.TwitterSearchScraper('"fire"').get_items(), 5000))
```

```python
df.head()
```

```python
df.shape
```

```python
df['user_location'] =  df['user'].apply(lambda x: x['location'])
df['user_location'].head(20)
```

```python
df['north_america']=[df['user_location'][i] if df['user_location'][i]=="North America" else "Nan" for i in range(df.shape[0])]
```

```python
df["north_america"].unique()
```

```python
df.shape
```

```python
df.head()
```
