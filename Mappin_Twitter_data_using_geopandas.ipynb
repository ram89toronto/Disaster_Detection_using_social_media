```python
location = pd.read_csv("best_model.csv")
location.head()
df_true = location[location["prediction"]==1]
df_true.head()
df_false = location[location["prediction"]==0]
df_false.head()
```

```python
longs =df_true["longitude"]
lats = df_true["latitude"]
```

```python
fig = plt.figure(num=None, figsize=(20,20) )
m = Basemap(projection='merc',llcrnrlat=-20,urcrnrlat=70,llcrnrlon=-190,urcrnrlon=-20,resolution='c')
m.drawcoastlines()
m.scatter(longs, lats, latlon = True, alpha =0.8)
m.fillcontinents(color ='tan', lake_color="blue", alpha =.2 )
m.drawparallels(np.arange(-90.,91.,30.),labels=[True,True,False,False],dashes=[2,2])
m.drawmeridians(np.arange(-180.,181.,60.),labels=[False,False,False,True],dashes=[2,2])
plt.title("Mercator Projection")
plt.show()
```
